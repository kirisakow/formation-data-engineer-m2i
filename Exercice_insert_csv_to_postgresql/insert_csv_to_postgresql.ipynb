# Lire les données

```python
import pandas as pd

df = pd.read_csv('c1202.csv', encoding='ISO-8859-1')

df
```

# Formater les données

```python
df.columns = ['secteur', 'sexe', '2005', '2006', '2007', '2008', '2009', '2010',
       '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
```

```python
df = pd.pivot_table(
    df,
    values=['2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'],
    index=['secteur', 'sexe']
)
df.reset_index(inplace=True)
df
```

# Insérer les données en base avec `SQLAlchemy`:

```python
from sqlalchemy import create_engine

uri_default = 'postgresql://postgres:Test1234@localhost:5432/db'
uri_psycopg2 = 'postgresql+psycopg2://postgres:Test1234@localhost:5432/db'

engine = create_engine(uri_psycopg2, echo=False)

df.to_sql(
    name='t_test',
    con=engine,
    if_exists='replace',
    index=False,
    chunksize=500,
)
```

# Résultat:

```python
engine.execute("SELECT * FROM t_test").fetchall()
```

# Une autre manière de se connecter à une DB Postgre (avec `psycopg2`):

```python
import psycopg2

conn = None
try:
    conn = psycopg2.connect(host='localhost', user='postgres', password='Test1234', database='db')
    cur = conn.cursor()
    cur.execute('SELECT version()')
    db_version = cur.fetchone()
    print(db_version)
    cur.close()
except (Exception, psycopg2.DatabaseError) as error:
    print(error)
finally:
    if conn is not None:
        conn.close()
        print('Database connection closed.')
```
